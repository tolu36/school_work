Question 1 – Using the Excel file dataA.xlsx, which contains a 500x3 data matrix (500 data points with 3
attributes), calculate both the mean and the covariance matrix.

```python
h_data =  pd.read_excel('dataA.xlsx', header=None)
d_array = np.array(h_data)
print('The mean vector of the 3 attributes are:')
print(np.mean(d_array.T, axis=1))
print('The covariance matrix of the 3 attributes are:')
print(np.cov(d_array.T))
```

Question 4 - From the Excel file document multinormal.xlsx, I have provided 10 examples of samples
drawn from a two-dimensional normal distribution (all in one file). In the data, there are 20 columns of
1000 samples. Example 1 is columns 1 and 2, example 2 is columns 3 and 4, etc. For each of the examples,
the mean is [0 0]T but the covariance matrix changes.
Plot the 2D points as a scatter plot for each example.
Calculate 10 separate covariance matrices. Plot the covariance value (the off-diagonal of the covariance
matrix) for each example. Similarly, plot the variance of the first column and the variance of the second
column for each example.
From the 2D scatter plots, what has changed from example to example? Think geometrically!
How might I have generated the covariance matrices that were used? Consider three things:  
1. What matrix transformation did I apply to change the covariance matrix for each example?  
2. What properties must a covariance have? (This one is trickier – one property is the matrix must
be symmetrical but there is one more important property...)  
3. How might one go about creating a covariance matrix that accomplishes the transformation
mentioned in point 1 and has the properties mentioned in point 2?  

```python
import seaborn as sn
h_data =  pd.read_excel('multNormal.xlsx', header=None)
for x, i in enumerate(list(range(h_data.shape[1]))[::2]):
    df = h_data.iloc[:,i:i+2]
    print(f"Example: {x+1}")
    plt.scatter(df.iloc[:,0], df.iloc[:,1], c = 'b', marker = '.')
    
    d_array = np.array(df)
    print(f'The covariance matrix for Example: {x+1}')
    cov_df = np.cov(d_array.T)
    plt.scatter(x=cov_df[0,0], y=0, c = 'r', marker = '.')
    plt.scatter(x=cov_df[0,1], y=cov_df[1,0], c = 'k', marker = '.')
    plt.scatter(x=0, y=cov_df[1,1], c = 'y', marker = '.')
    plt.show()
    print(cov_df)
    sn.heatmap(cov_df, annot=True, fmt='g')
    plt.show()
```

(I plotted a covariance matrix, to my understanding of the question that was what was asked. Though I plotted the covariance and variance points directly on 2D plots of the examples as well.)

The noticeable change that we see example to example is that the 2 columns start off practically uncorrelated and by example 6 they have strong positive correlation, however, this example they begin to be less correlated til the last example were they are practically uncorrelated as they started off. We also see that the variance of column 1 (left column of each example) begins to decrease while the column 2 variance increases, however, the two columns are most correlated when their variances are the closest together.  Geometrically this looks like a saddle point where the variance tells the spread of the data on the axes and the covariance tells us the orientation (positive or negative).  
1. A linear transformation is applied to the original data ie (example 1) which leads to a rotated and scaled covariance matrices for all the other examples. 
2. A covariance matrix must be positive semi-definite and all variance values are real non-negative numbers. This is the reason for the saddle shape, because there is no actual minimum point. 
3. We can create the covariance matrices found in these examples by transforming the initial dataset (example 1) with a scaling and rotating matrix. Thus, Y = RSX where R is the rotating matrix [[cos(\theta), -sin(\theta)],[sin(\theta), cos(\theta)]] and S is the scale matrix [[s_x,0], [0,s_y]]
